```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X =np.random.rand(100,2) #100個亂數,每個亂數取兩個值
```

```python
X[0]
```

```python
plt.scatter(X[:,0],X[:,1], s=50)
```

### 上面的圖如何分為3類?

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3) #需要分為三類
```

```python
clf.fit(X)
```

```python
clf.labels_ #分類結果
```

```python
plt.scatter(X[:,0], X[:,1], c = clf.labels_)
```
