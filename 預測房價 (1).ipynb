```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.feature_names
```

```python
X = boston.data
Y = boston.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,
                                                   test_size=0.3,
                                                   random_state = 123)
```

```python
len(X)
```

```python
len(y_train)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
y_predict = regr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict) #difference between real and predict
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

[小技巧] 善用 enumerate

```python
L = ("A","B","C")
```

```python
for i in L:
    print(i)
```

```python
for i in range(3):
    print(i+1, L[i])
```

```python
list(enumerate(L)) #代替以上的功能
```

```python
for i in enumerate(L): #直接帶入enumerate,但是格式不是我要的
    print(i)
```

```python
for i ,s in enumerate(L): #帶入想要的格式
    print(i+1, s)
```

#### [小技巧] 畫多個圖

```python
x = np.linspace(-10,10,100)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
#如果不想畫在一起怎麼辦
```

```python
plt.subplot(2,2,1) #圖形排法為,2X2,的第一張圖
plt.plot(x,np.sin(x))

plt.subplot(2,2,2) #圖形排法為,2X2,的第2張圖
plt.plot(x,np.cos(x))

plt.subplot(2,2,3) #圖形排法為,2X2,的第3張圖
plt.plot(x,2+x**2)

plt.subplot(2,2,4) #圖形排法為,2X2,的第4張圖
plt.plot(x,10+3*x**3)
```

```python
plt.figure(figsize=(8,10)) #指定圖形大小xy
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:,i], Y, s=1)
    plt.ylabel("price")
    plt.xlabel(feature)
    plt.tight_layout() #用這行指令避免標籤擠在一起
```
